```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file using the full file path
df = pd.read_csv('C:/Users/gaelt/Downloads/sales_data new.csv')

# Preview the first few rows
print("Preview of the sales data:")
print(df.head())

# Analyze Most Prevalent Products
# Find the most prevalent products
product_counts = df['Product Name'].value_counts()
print("\nTop 10 Most Purchased Products:")
print(product_counts.head(10))

# Analyze Large Baskets
# Calculate the basket sizes
basket_sizes = df.groupby('OrderID')['Product Name'].count()

# Define a large basket threshold (e.g., > 2 items)
large_basket_threshold = 2
large_basket_orders = basket_sizes[basket_sizes > large_basket_threshold]

print("\nLarge Basket Orders (more than 2 items):")
print(large_basket_orders)

# Stores with Large-Basket Buyers
# Filter for large-basket orders
large_basket_stores = df[df['OrderID'].isin(large_basket_orders.index)]

# Group by StoreID and count unique large-basket orders
store_counts = large_basket_stores.groupby('StoreID')['OrderID'].nunique()

print("\nStores with Large Basket Buyers:")
print(store_counts)

# Visualize Top Stores
# Plot a bar chart for store counts
print("\nVisualizing Top Stores with Large-Basket Customers...")
store_counts.sort_values(ascending=False).plot(kind='bar', figsize=(8, 6))
plt.title("Top Stores for Large-Basket Customers")
plt.xlabel("Store ID")
plt.ylabel("Number of Large Baskets")
plt.show()

# Top Products for Large Baskets
# Find top products in large-basket orders
top_large_basket_products = large_basket_stores['Product Name'].value_counts().head(5)

print("\nTop 5 Products in Large Baskets:")
print(top_large_basket_products)

# Categorize Customer Demographics
# Find the percentage makeup of products in large baskets
category_makeup = large_basket_stores['Product Name'].value_counts(normalize=True) * 100

print("\nCategory Makeup of Large Baskets (%):")
print(category_makeup)

# Visualize Product Categories
# Plot a pie chart for category makeup
print("\nVisualizing Product Category Distribution in Large Baskets...")
category_makeup.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Product Category Distribution in Large Baskets")
plt.ylabel("")  # Remove the y-axis label
plt.show()
```

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file using the full file path
df = pd.read_csv('C:/Users/gaelt/Downloads/sales_data new.csv')

# Preview the first few rows
print("Preview of the sales data:")
print(df.head())

# Analyze Most Prevalent Products
# Find the most prevalent products
product_counts = df['Product Name'].value_counts()
print("\nTop 10 Most Purchased Products:")
print(product_counts.head(10))

# Analyze Large Baskets
# Calculate the basket sizes
basket_sizes = df.groupby('OrderID')['Product Name'].count()

# Define a large basket threshold (e.g., > 2 items)
large_basket_threshold = 2
large_basket_orders = basket_sizes[basket_sizes > large_basket_threshold]

print("\nLarge Basket Orders (more than 2 items):")
print(large_basket_orders)

# Summary of large basket frequency
print(f"\nNumber of large baskets: {len(large_basket_orders)}")
print(f"Percentage of large baskets: {len(large_basket_orders) / len(basket_sizes) * 100:.2f}%")

# Stores with Large-Basket Buyers
# Filter for large-basket orders
large_basket_stores = df[df['OrderID'].isin(large_basket_orders.index)]

# Group by StoreID and count unique large-basket orders
store_counts = large_basket_stores.groupby('StoreID')['OrderID'].nunique()

print("\nStores with Large Basket Buyers:")
print(store_counts)

# Summary of top stores
top_store = store_counts.idxmax()
print(f"\nStore with the most large-basket buyers: {top_store} ({store_counts[top_store]} orders)")

# Visualize Top Stores
# Plot a bar chart for store counts
print("\nVisualizing Top Stores with Large-Basket Customers...")
store_counts.sort_values(ascending=False).plot(kind='bar', figsize=(8, 6))
plt.title("Top Stores for Large-Basket Customers")
plt.xlabel("Store ID")
plt.ylabel("Number of Large Baskets")
plt.show()

# Top Products for Large Baskets
# Find top products in large-basket orders
top_large_basket_products = large_basket_stores['Product Name'].value_counts().head(5)

print("\nTop 5 Products in Large Baskets:")
print(top_large_basket_products)

# Categorize Customer Demographics
# Find the percentage makeup of products in large baskets
category_makeup = large_basket_stores['Product Name'].value_counts(normalize=True) * 100

print("\nCategory Makeup of Large Baskets (%):")
print(category_makeup)

# Visualize Product Categories
# Plot a pie chart for category makeup
print("\nVisualizing Product Category Distribution in Large Baskets...")
category_makeup.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title("Product Category Distribution in Large Baskets")
plt.ylabel("")  # Remove the y-axis label
plt.show()

# Export analysis results to an Excel file
output_file = "sales_data_analysis.xlsx"

# Create a writer object for writing multiple DataFrames to one Excel file
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    # Analyze Best-Selling Items by Store
    store_group = df.groupby('StoreID')['Product Name'].value_counts()
    store_group.to_excel(writer, sheet_name='Best Selling Items by Store')

    # Analyze Overall Best Sellers
    overall_best_sellers = df['Product Name'].value_counts()
    overall_best_sellers.to_excel(writer, sheet_name='Best Sellers Overall')

    # Save large basket orders
    large_basket_orders.to_excel(writer, sheet_name='Large Basket Orders')

    # Save category analysis
    category_makeup.to_excel(writer, sheet_name='Category Analysis')

print(f"Analysis results have been saved to {output_file}.")
```
